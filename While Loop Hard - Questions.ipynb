```python
1. Write a Python program to find the least common multiple (LCM) of two given integers.
    Data:
        lcm(4, 6)
        lcm(15, 17)
    Expected Output:
        12
        255
        
        
        
        
```

```python
def lcm(a, b):
    if a > b:
        greater = a
    else:
        greater = b
    while True:
        if (greater % a == 0) and (greater % b == 0):
            lcm = greater
            break
        greater += 1
    return lcm

print("LCM of 4 and 6 is:", lcm(4, 6))
print("LCM of 15 and 17 is:", lcm(15, 17))
```

```python
def sum_of_cubes(n):
    if n <= 0:
        return "Input must be a positive integer"
    sum_cubes = sum(i**3 for i in range(1, n))
    return sum_cubes
data = 3
result = sum_of_cubes(data)
print(f"Sum of cubes: {result}")
```

```python
4. Write a Python program to find unique triplets whose three elements gives the sum of zero from an array of n integers.
    Data:
        x = [1, -6, 4, 2, -1, 2, 0, -2, 0 ]
    Expected Output:
        [(-6, 2, 4)]
        
        
        
        
```

```python
def find_triplets(arr):
    arr.sort()
    n = len(arr)
    triplets = []

    for i in range(n - 2):
        if i > 0 and arr[i] == arr[i - 1]:
            continue
        
        left, right = i + 1, n - 1
        while left < right:
            total = arr[i] + arr[left] + arr[right]
            if total == 0:
                triplets.append((arr[i], arr[left], arr[right]))
                while left < right and arr[left] == arr[left + 1]:
                    left += 1
                while left < right and arr[right] == arr[right - 1]:
                    right -= 1
                left += 1
                right -= 1
            elif total < 0:
                left += 1
            else:
                right -= 1
    return triplets
x = [1, -6, 4, 2, -1, 2, 0, -2, 0]
result = find_triplets(x)
print(result)
```

```python
5. Write a Python program to find the value of n where n degrees of number 2 are written sequentially in a line without spaces.
    Data:
        "2481632"
        "248163264"
    Expected Output:
        5
        6
        
        
        
        
```

```python
def find_n_for_powers_of_2(data):
    n = 1
    concatenated_powers = ""
    while True:
        concatenated_powers += str(2 ** n)
        if concatenated_powers == data:
            return n
        elif len(concatenated_powers) > len(data):
            return -1
        n += 1
data1 = "2481632"
data2 = "248163264"

print(find_n_for_powers_of_2(data1))
print(find_n_for_powers_of_2(data2))
```

```python
6. Write a Python program that accept a positive number and subtract from this number the sum of its digits and so on. 
   Continues this operation until the number is positive.
    Data:
        9
        21
    Expected Output:
        1
        3
        
        
        
        
```

```python
n = int(input("Enter a positive number: "))
while n > 0:
    digits_sum = 0
    temp = n
    while temp > 0:
        digits_sum += temp % 10
        temp //= 10
    prev_n = n
    n = n - digits_sum

print(prev_n)
```

```python
7. Write a Python program to compute the summation of the absolute difference of all distinct pairs in an given array 
   (non-decreasing order).
    Data:
        [1, 2, 3]
        [1,4,5]
    Expected Output:
        4
        8
        
        
        
```

```python
def sum_of_absolute_differences(arr):
    n = len(arr)
    total_sum = 0
    for i in range(n):
        for j in range(i + 1, n):
            total_sum += abs(arr[i] - arr[j])
    return total_sum
data1 = [1, 2, 3]
data2 = [1, 4, 5]
print(sum_of_absolute_differences(data1))
print(sum_of_absolute_differences(data2))
```

```python
8. Write a Python program to find common divisors between two numbers in a given pair.
    Data:
        (2, 4)
        (2, 8)
        (12, 24)
    Expected Output:
        Number of common divisors:  2
        Number of common divisors:  2
        Number of common divisors:  6
            
            
            
            
```

```python
def common_divisors_count(a, b):
    def gcd(x, y):
        while y:
            x, y = y, x % y
        return x

    gcd_ab = gcd(a, b)
    count = 0
    for i in range(1, gcd_ab + 1):
        if gcd_ab % i == 0:
            count += 1
    
    return count
pairs = [(2, 4), (2, 8), (12, 24)]
for pair in pairs:
    a, b = pair
    result = common_divisors_count(a, b)
    print(f"Number of common divisors: {result}")
```

```python
9. Write a Python program to reverse the digits of a given number and add it to the original,If the sum is not a palindrome 
   repeat this procedure.
    Data:
        1234
        1473
        121
    Expected Output:
        5555
        9339
        121
        
        
        
```

```python
def is_palindrome(num):
    return str(num) == str(num)[::-1]

def reverse_number(num):
    return int(str(num)[::-1])

def reverse_and_add_until_palindrome(num):
    while not is_palindrome(num):
        reversed_num = reverse_number(num)
        num += reversed_num
    return num
numbers = [1234, 1473, 121]
for number in numbers:
    result = reverse_and_add_until_palindrome(number)
    print(result)
```

```python
10. Write a Python program to find the maximum sum of a contiguous subsequence from a given sequence of 
    numbers a1, a2, a3, ... an. A subsequence of one element is also a continuous subsequence.
    Expected Output:
        Input number of sequence of numbers you want to input (0 to exit):
        2
        Input numbers:
        6
        4
        Maximum sum of the said contiguous subsequence:
        10
        Input number of sequence of numbers you want to input (0 to exit):
        0
        
        
        
        
```

```python
def max_subarray_sum(nums):
    if not nums:
        return 0

    max_current = max_global = nums[0]

    for num in nums[1:]:
        max_current = max(num, max_current + num)
        if max_current > max_global:
            max_global = max_current

    return max_global

def main():
    while True:
        try:
            n = int(input("Input number of sequence of numbers you want to input (0 to exit): "))
            if n == 0:
                break
            if n < 0:
                print("Please enter a positive number or 0 to exit.")
                continue

            nums = []
            print("Input numbers:")
            for _ in range(n):
                nums.append(int(input()))

            max_sum = max_subarray_sum(nums)
            print("Maximum sum of the said contiguous subsequence:")
            print(max_sum)
        except ValueError:
            print("Invalid input. Please enter integer values.")

if __name__ == "__main__":
    main()
```

```python
11. Write a Python program that reads n digits (given) chosen from 0 to 9 and prints the number of combinations 
    where the sum of the digits equals to another given number (s). Do not use the same digits in a combination. 
    Expected Output:
        Input number of combinations and sum, input 0 0 to exit:
        4 6
        2 4
        0 0
        2
        
        
        
        
```

```python
import itertools

def count_combinations(n, s):
    digits = list(range(10))
    count = 0
    for combination in itertools.combinations(digits, n):
        if sum(combination) == s:
            count += 1
    return count

def main():
    while True:
        user_input = input("Input number of combinations and sum, input 0 0 to exit:\n")
        n, s = map(int, user_input.split())
        if n == 0 and s == 0:
            break
        result = count_combinations(n, s)
        print(result)

if __name__ == "__main__":
    main()
```

```python
12. Write a Python program to compute the sum of first n given prime numbers.
    Expected Output:
        Input a number (n≤10000) to compute the sum:(0 to exit)
        50
        Sum of first 50 prime numbers:
        5117
        3456
        Sum of first 3456 prime numbers:
        52238131
        0
        
        
        
        
```

```python
import itertools

def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

def sum_of_primes(n):
    prime_sum = 0
    prime_count = 0
    num = 2  # Start checking for prime from the first prime number
    while prime_count < n:
        if is_prime(num):
            prime_sum += num
            prime_count += 1
        num += 1
    return prime_sum

def main():
    while True:
        try:
            n = int(input("Input a number (n<=10000) to compute the sum:(0 to exit)\n"))
            if n == 0:
                break
            elif n > 10000:
                print("Please enter a number less than or equal to 10000.")
                continue
            print(f"Sum of first {n} prime numbers:")
            print(sum_of_primes(n))
        except ValueError:
            print("Please enter a valid integer.")

if __name__ == "__main__":
    main()
```

```python
13. Write a Python program to restore the original string by entering the compressed string with this rule. However, 
    the character does not appear in the restored character string. 
    Data:
        XY#6Z1#4023
        #39+1=1#30
    Expected Output:
        Original text: XY#6Z1#4023
        XYZZZZZZ1000023
        Original text: #39+1=1#30
        999+1=1000
        
        
        
        
```

```python
def restore_string(compressed_str):
    original_str = compressed_str
    restored_str = ""
    i = 0
    
    while i < len(compressed_str):
        if compressed_str[i] == '#':
            # Get the number after '#'
            j = i + 1
            num_str = ""
            while j < len(compressed_str) and compressed_str[j].isdigit():
                num_str += compressed_str[j]
                j += 1
            
            # Convert the number string to an integer
            num = int(num_str)
            
            # Append the character num times to the restored string
            restored_str += str(num) * num
            
            # Move the index to the next character after the number
            i = j
        else:
            restored_str += compressed_str[i]
            i += 1
            
    return original_str, restored_str

def main():
    data = [
        "XY#6Z1#4023",
        "#39+1=1#30"
    ]
    
    for compressed_str in data:
        original_str, restored_str = restore_string(compressed_str)
        print(f"Original text: {original_str}")
        print(restored_str)

if __name__ == "__main__":
    main()
```

```python
14. Write a Python program to find the number of combinations that satisfy p + q + r + s = n where n is a given 
    number <= 4000 and p, q, r, s in the range of 0 to 1000. 
    Expected Output:
        Input a positive integer: (ctrl+d to exit)
        343
        Number of combinations of a,b,c,d: 6843880
        d
        
        
        
        
```

```python
def count_combinations(n):
    count = 0
    for p in range(0, min(n, 1000) + 1):
        for q in range(0, min(n - p, 1000) + 1):
            for r in range(0, min(n - p - q, 1000) + 1):
                s = n - p - q - r
                if 0 <= s <= 1000:
                    count += 1
    return count

def main():
    try:
        while True:
            n = int(input("Input a positive integer: (ctrl+d to exit)\n"))
            if n < 0:
                print("Please enter a positive integer.")
                continue
            if n > 4000:
                print("Please enter a number less than or equal to 4000.")
                continue
            result = count_combinations(n)
            print(f"Number of combinations of p, q, r, s: {result}")
    except EOFError:
        print("Exiting...")

if __name__ == "__main__":
    main()
```

```python
15. Write a Python program which adds up columns and rows of given table as shown in the specified figure.
    Input number of rows/columns (0 to exit)
    Data:
        Input cell value:
        34 86 45 56
        12 23 45 67
    Expected Output:
        Input number of rows/columns (0 to exit)
        2
        Input cell value:
        34 86 45 56
        12 23 45 67
        Result:
        34   86   45
        12   23   45
        46  109   90
        Input number of rows/columns (0 to exit)
        0
        
        
        
        
```

```python
def print_table_with_sums(table):
    # Calculate the sum of each row
    row_sums = [sum(row) for row in table]

    # Calculate the sum of each column
    column_sums = [sum(column) for column in zip(*table)]

    # Print the table with row sums
    for row in table:
        print(" ".join(f"{cell:4}" for cell in row))

    # Print the sums
    print(" ".join(f"{sum:4}" for sum in column_sums))
    print(" ".join(f"{sum:4}" for sum in row_sums))

def main():
    while True:
        # Ask for the number of rows/columns
        n = int(input("Input number of rows/columns (0 to exit): "))
        if n == 0:
            break

        # Read the table data
        table = []
        for i in range(n):
            row = list(map(int, input("Input cell value: ").split()))
            table.append(row)

        print("Result:")
        print_table_with_sums(table)

if __name__ == "__main__":
    main()
```

```python
16. Write a Python program to calculate Euclid's totient function of a given integer. Use a primitive method to 
    calculate Euclid's totient function.
    Data:
        (10)
        (15)
        (33)
    Expected Output:
        4
        8
        20
        
        
        
        
```

```python
def gcd(a, b):
    while b:
        a, b = b, a % b
    return a

def euler_totient(n):
    count = 0
    for i in range(1, n + 1):
        if gcd(n, i) == 1:
            count += 1
    return count

# Test cases
test_data = [10, 15, 33]
expected_output = [4, 8, 20]

for num, expected in zip(test_data, expected_output):
    result = euler_totient(num)
    print(f"Euler's Totient function φ({num}) = {result}, Expected Output = {expected}")
```

```python
import math

def are_coprime(a, b):
    return math.gcd(a, b) == 1

# Test data
test_data = [
    (17, 13),
    (15, 21),
    (25, 45)
]

# Expected Output:
# True
# True
# False

# Check coprimeness for each pair in the test data
for a, b in test_data:
    print(are_coprime(a, b))
```

```python
18. Write a Python program to print a given N by M matrix of numbers line by line in forward > backwards > forward >... order.
    Data:
        Input matrix:
            [[1, 2, 3,4],
             [5, 6, 7, 8],
             [0, 6, 2, 8],
             [2, 3, 0, 2]]
    Expected Output:
        1
        2
        3
        4
        8
        7
        6
        5
        0
        6
        2
        8
        2
        0
        3
        2
        
        
        
        
```

```python
def print_matrix_in_zigzag_order(matrix):
    for i, row in enumerate(matrix):
        if i % 2 == 0:
            for num in row:
                print(num)
        else:
            for num in reversed(row):
                print(num)

# Input matrix
matrix = [
    [1, 2, 3, 4],
    [5, 6, 7, 8],
    [0, 6, 2, 8],
    [2, 3, 0, 2]
]

# Print the matrix in zigzag order
print_matrix_in_zigzag_order(matrix)
```

```python
19. Write a Python program to remove all instances of a given value from a given array of integers and find the length of 
    the new array.
    Data:
        ([1, 2, 3, 4, 5, 6, 7, 5], 5)
        ([10,10,10,10,10], 10)
        ([10,10,10,10,10], 20)
        ([], 1)
    Expected Output:
        6
        0
        5
        0
        
        
        
        
```

```python
def remove_value_and_get_length(arr, value):
    # Use list comprehension to filter out the given value
    new_arr = [x for x in arr if x != value]
    # Return the length of the new array
    return len(new_arr)

# Test cases
data = [
    ([1, 2, 3, 4, 5, 6, 7, 5], 5),
    ([10, 10, 10, 10, 10], 10),
    ([10, 10, 10, 10, 10], 20),
    ([], 1)
]

# Expected Output: 6, 0, 5, 0
for arr, value in data:
    print(remove_value_and_get_length(arr, value))
```
